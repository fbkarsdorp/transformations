### Dataset

In order to quicly obtain a fairly large collection of magical transformations in folktales, we came up with the following data collection and annotation strategy. First we retrieved all texts from the Dutch Folktale Database that matched one of our seed queries. These seed queries are the following:

- all texts matching the string `*verander*` (either on the text or on the keywords given);
- all texts matching `*gedaante*` (either on the text or on the keywords given); 
- all texts matching `*metamorf*` (either on the text or on the keywords given) and
- all texts matching `*transform*` (either on the text or on the keywords given).

These resulted in about 750 folktales. All stories in this set were than manually annotated using the following annotation schema. For each magical transformation in a text we notate who or what is the source of the transformation, who or what is the target of the transformation and (if applicable) the verb describing the transformation. We assign both the source and the target to a category in the following hierarchical ontological tree:

### Detecting motifs of transformation in folk tales

The remainder of this notebook describes some preliminary experiments to detect motifs of transformation in folktales. We make use of the initial manually annotated dataset containing about 700 instances of magical transformations. Each sentence in about 900 documents, was tagged with either `transformation` or `no-transformation`. The task is to develop a system that rapidly can deliver more instances of transformations. This system need not be perfect. It should act as a helping hand in extracting more instances of magical transformations to perform our analyses on.

First we import all the necessary packages and modules into the workspace.

```python
import codecs
from collections import defaultdict
import random
import sqlite3 as sqlite
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn import cross_validation
from sklearn.cross_validation import KFold
from sklearn.metrics import classification_report, f1_score, precision_recall_curve, auc
from sklearn.metrics import confusion_matrix
from sklearn.svm import LinearSVC
from sklearn.linear_model import SGDClassifier, LogisticRegression, RidgeClassifier
from sklearn.linear_model import Perceptron, PassiveAggressiveClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.pipeline import FeatureUnion, Pipeline
from sklearn.feature_selection import chi2, SelectKBest, f_classif

from pystruct.models import ChainCRF
from pystruct.learners import OneSlackSSVM

from gensim.models.word2vec import Word2Vec

from ssplit import split
from plm import ParsimoniousLM
```

Let's start with reading in the dataset. The dataset is formatted as a single file where each line contains a sentence from a particular story, aloong with the idnumber of the corresponding story and a tag representing whether a magical transformation occurs in that sentences.

```python
idnumbers = []
ids = set()
training_sentences, labels = [], []
for line in codecs.open("../data/transformation_annotated-2.txt", "r", "utf-8"):
    idnumber, tag, sentence = line.strip().split('|', 2)
    if sentence.strip():
        idnumbers.append(idnumber.replace('.txt', ''))
        ids.add(idnumber.replace('.txt', ''))
        labels.append(-1 if tag == 'O' else 1)
        training_sentences.append(sentence)
```

### Character $n$-grams

I will explore a range of different feature representations. As a first baseline I will represent sentences as bags of character ngrams. I make use of the scikit-learn toolkit to extract the features. The same toolkit is used to develop a classification system.

```python
vectorizer = TfidfVectorizer(ngram_range=(4, 4), min_df=2, max_df=0.6, analyzer='char')
X = vectorizer.fit_transform(training_sentences)
y = np.array(labels)
```

In order to be able to efficiently compare different classification systems, we write a small function `evaluate` that takes as argument a classifier and performs 10-fold cross-validation on the dataset.

```python
def evaluate(clf, X, y, n_folds=10):
    scores = np.zeros(n_folds)
    for k, (train, test) in enumerate(KFold(len(y), n_folds=n_folds, shuffle=True)):
        X_train, X_test, y_train, y_test = X[train], X[test], y[train], y[test]
        clf.fit(X_train, y_train)
        preds = clf.predict(X_test)
        print classification_report(y_test, preds)
        scores[k] = f1_score(y_test, preds, average= 'macro')
    print "F1: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2)
```

The first classifier I'm going to experiment with is a Linear SVC. Let's first do a basic run and see what comes out... Note that since the class labels are heavily unbalanced, we must set the `class_weight` parameter to `auto`.

```python
evaluate(LinearSVC(class_weight='auto'), X, y)
```

The results show us that although the transformation class is a very small minority class, the system is still able to fairly accurately extract instances of transformation. However, it will be clear that there is also enough room left for improvement. 

One challenging aspect of SVMs is to optimize their parameters. However, I will first focus on developing and experimenting with different feature representations as most performance improvement can be gained from that. Before we do that, we want to make sure that SVMs are not the only well preforming classification system. I therefore compare the results to those of a Stochastic Gradient Descent Classifier as implemented in the scikit-learn toolkit.

```python
evaluate(SGDClassifier(alpha = 0.0001, n_iter=np.ceil(10**6 / len(y)), 
                       class_weight='auto'), X, y)
```

The results are very similar to those obtained with the Linear SVC. For now we will stick to the SVC.

```python
n_features = np.arange(0.1, 1.05, 0.05) * X.shape[1]
def evaluate_features(n_features)
    scores = np.zeros(n_features.shape[0])
    for i, size in enumerate(n_features):
        k_scores = np.zeros(10)
        for k, (train, test) in enumerate(KFold(len(y), n_folds=10, shuffle=True, random_state=0)):
            X_train, X_test, y_train, y_test = X[train], X[test], y[train], y[test]
            ch2 = SelectKBest(chi2, k=int(size))
            X_train = ch2.fit_transform(X_train, y_train)
            X_test = ch2.transform(X_test)
            clf = LinearSVC(class_weight='auto')
            clf.fit(X_train, y_train)
            preds = clf.predict(X_test)
            k_scores[k] = f1_score(y_test, preds, average='macro')
        scores[i] = k_scores.mean()
    return pd.Series(scores, index=n_features)
```

### Active Learning

Manually annotating data collections often is a highly costly endeavor. This holds especially true in the Humanities where scholars deal with high dimensional, complex data. Automatic systems from Machine Learning or Information Retrieval can assist annotation projects. Typically these systems need to be trained on a set of annotated data on the basis of which they ‘learn’ how to annotate new, unseen data. Automatically obtained annotations can then be evaluated by human annotators. This co-operation between human and machine greatly reduces the intense cost of manually annotating data. A common characteristic of these automatic systems is that they require large amounts of training data to be put to action reliably. A famous quote by Mercer’s reads “There is no data like more data”. It addresses the issue that in many Machine Learning problems more data is often more important than better algorithms. Notwithstanding this adagium, obtaining such large amounts of annotated data is often not feasible, at least, not without the help of a machine. To reduce the cost of a brute-force data collection struggle, we propose an iterative data annotation loop which tries to optimize the learning curve of a learning system. Key feature of this loop is that it makes a ranking of the order in which new items should be annotated to more efficiently arrive at a well-preforming system. Most modern Machine Learning systems (e.g. Support Vector Machines) utilize some kind of decision function which is used to classify objects into different classes. The values of this decision function can be seen as representations of the certainty with which the system made its classifications. We show that by focussing the annotation endeavor first and foremost on those cases where the learning system exhibits the highest classification *uncertainty*, we are able to develop a better learning system with less annotation effort.



The results from the previous analysis show that most of the errors made by the system fall within the range $-0.8 > n < 0$. The following density plot illustrates this clearly:

```python
d = pd.Series([pr for pr, t, p in zip(probs, y_test, preds) if 
               (t == 1 and p == -1) or (t == -1 and p == 1)])
sns.distplot(d, hist=False, rug=True)
```

The following experiments aim to provide some empirical evidence for the idea that by feeding a classifier with exemplars it is most uncertain about, we can achieve the same performance using less hand-labeled items. 

We divide the original dataset into a test and training set. We will use different chunk sizes of the training set to predict the outcome of the test set. 

```python
split = int(0.4 * len(training_sentences))
TRAIN, TEST = training_sentences[:split], training_sentences[split:]
TRAIN_Y, TEST_Y = labels[:split], labels[split:]
```

The following functions are used to compare different data feeding schemas. `linear_select` simply takes the data as it is given, `random_select` shuffles the data before it is given to the classifier and `uncertainty_select` selects the exemplars the classifier is most uncertain about.

```python
def learning_curve(X, y, X_TEST, y_TEST, clf, select_fn):
    scores = []
    for size in np.arange(0.1, 1.05, 0.05):
        # first select a portion of data to train the model
        split = int(size * len(X))
        X_sents, X_rest = X[:split], X[split:]
        y_labels, y_rest = y[:split], y[split:]
        # fir and transform the data to a character based tf-idf ngram model
        vec = TfidfVectorizer(ngram_range=(2,7), min_df=2, max_df=0.6, analyzer='char')
        X_train = vec.fit_transform(X_sents)
        y_train = np.array(y_labels)
        # fit the classifier
        clf.fit(X_train, y_train)
        # fit the test data
        X_test = vec.transform(X_TEST)
        y_test = np.array(y_TEST)
        # make a prediction based on the trained classifier
        pred = clf.predict(X_test)
        # append the score to the score list
        scores.append(f1_score(y_test, pred, average= 'macro'))
        # select the training data for the next round
        if X_rest and y_rest:
            X, y = select_fn(X_sents, X_rest, y_labels, y_rest, clf=clf, vec=vec)
    return pd.Series(scores)
        
def random_select(X_sents, X_rest, y_labels, y_rest, **kwargs):
    rest = zip(X_rest, y_rest)
    random.shuffle(rest)
    X_rest, y_rest = zip(*rest)
    return X_sents + list(X_rest), y_labels + list(y_rest)

def linear_select(X_sents, X_rest, y_labels, y_rest, **kwargs):
    return X_sents + X_rest, y_labels + y_rest

def uncertainty_select(X_sents, X_rest, y_labels, y_rest, clf, vec):
    decisions = clf.decision_function(vec.transform(X_rest))
    m = pd.Series(decisions).mean()
    decisions = sorted(range(len(decisions)), key=lambda i: abs(m - decisions[i]))
    for i in decisions:
        X_sents.append(X_rest[i])
        y_labels.append(y_rest[i])
    return X_sents, y_labels  
```

```python
clf = LinearSVC(class_weight='auto', random_state=0)
random_curve = learning_curve(TRAIN, TRAIN_Y, TEST, TEST_Y, clf, random_select)
linear_curve = learning_curve(TRAIN, TRAIN_Y, TEST, TEST_Y, clf, linear_select)
uncertainty_curve = learning_curve(TRAIN, TRAIN_Y, TEST, TEST_Y, clf, uncertainty_select)
```

```python
chunks = pd.Series([int(size * len(TRAIN)) for size in np.arange(0.1, 1.05, 0.05)])
df = pd.concat([random_curve, linear_curve, uncertainty_curve], axis=1,
               keys = ['random', 'linear', 'uncertain'])
df.index = chunks
df.plot()
```

The graph above illustrates beautifully the strength of feeding data to a classifier in a non-random way.

### Structured prediction?

```python
graph = defaultdict(list)
lastid = None
id = 0
sent_id = 0
for line in codecs.open("../data/transformation_annotated-2.txt", "r", "utf-8"):
    idnumber, tag, sentence = line.strip().split('|', 2)
    if lastid is None or lastid != idnumber:
        lastid = idnumber
        id += 1
    if sentence.strip():
        graph[id].append(sent_id)
        sent_id += 1
```
